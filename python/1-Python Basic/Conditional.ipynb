```python
## LOOp CONTROL S?TATEMENT 

#1. break - exits the loop prematurely

for i in range(10):
    if i == 5:
        break
    print(i)
```

```python
## 2. continue 
#  -skips the current iteration and continuess with the next




for i in range(10):
    if i%2 ==0:
        continue
    
    print(i)
```

```python
## 3. pass - null statement or operation , it does norhing

for i in range(5):
    if i == 3:
        pass
    print(i)
        

```

```python
# Example 
# Calculate the sum of first N natural numbers using a while anf for loop



num = int(input("Enter the Natural Number: "))
sum = 0
count = 1

while count <= num:
    sum = count + sum
    count = count +1
print(sum)
```

```python
## USing for loop

num = 6

sum = 0

for i in range(num+1):
    sum = sum + i
    
print(sum)
```
