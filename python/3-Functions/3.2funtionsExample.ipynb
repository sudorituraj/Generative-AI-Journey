```python
# 1. Temperature Conversion



def convert_temp(temp,unit):
    if unit =="C":
        return temp * 9/5+32
    
    elif unit == "F":
        return(temp-32)*5/9
    
print(convert_temp(25,"C"))
print(convert_temp(77,"F"))
    
```

```python
#### 2. Srtong Password

def is_strong_pass(pwd):
    if len(pwd)<8:
        return False
    
    if not any(char.isdigit() for char in pwd):
        return False
    if not any(char.islower() for char in pwd):
        return False
    
    if not any(char.isupper() for char in pwd):
        return False
    if not any(char in "!@#$%^&()_+ " for char in pwd):
        return False
    
    return True


print(is_strong_pass("WSDFrjur"))
print(is_strong_pass("Ritu143@#"))
    
    
```

```python

#### 3. calculate the total cost of items in a shopping Cart


def calculate_total_cost(cart):
    
    total_cost= 0
    
    for item in cart:
        total_cost += item["price"] * item["quantity"]
        
    return total_cost
        
cart =[
    {"name":"Apple","price":0.4,"quantity":4},
    {"name":"Banana","price":0.6,"quantity":6},
    {"name":"Orange","price":0.2,"quantity":3}
    
]


## function calling


total_cost = calculate_total_cost(cart)
print("Total cost of the cart:",total_cost)


```

```python
##### 4. Check IF a string is palindrome

def str_palindrome(text):
    
    
    
    cleaned = text.replace(" ","").lower()
    return cleaned == cleaned[::-1]
    
text = "madam"
    
if str_palindrome(text):
        print(f"{ text} is a palindrome")
        
else:
        print(f"{text} is not a palindron")
    

   
    
```

```python
#### 5. Cal the factoricals of a number using recursion


def factorial(n):
    if n == 0:
        return 1
    else:
        return n * factorial(n-1)
    
print("Here is the factoriqal of the given number: ",factorial(4))



```

```python


def count_word(file_path):
    word_count = {}
    with open(file_path, "r") as file:
        for line in file:
            words = line.split()  # fixed typo here
            
            for word in words:
                word = word.lower().strip(',.!?:;"')
                word_count[word] = word_count.get(word, 0) + 1

    return word_count  # <- also add return to get result

filepath = 'sample.txt'
word_freq = count_word(filepath)
print(word_freq)
```
