```python
### Deafult parameter

def greet(name ="Guest"):
    print(f"Hello {name} Welcome to the Paradise")
    
greet()
greet("Ritu")
```

```python
#### Variable length Agrumnets

## Positional  Arguments

print("Postitional Arguments ")

def print_num(*args):
    for num in args:
        print(num)
        
print_num(1,2,3,4,5,6,7,8,"Ritu")


### Keywords Arguments
print("Keywords Argments: ")

def print_details(**kwargs):
    for keys,value in kwargs.items():
        print(f"{keys}:{value}")
        
print_details(name="Ritu",age=32,country ="India")
        
```

```python


def print_details(*args,**kwargs):
    
    for val in args:
        print(f"Postitional Arguments: {val}")
    
    for keys,values in kwargs.items():
        print(f"{keys}:{values}")
        

print_details(1,2,3,4,"krish",name="Ritu",age=32,country ="India")


```
