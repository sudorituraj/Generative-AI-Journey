```python
### iterating Over Dictionaries

## Iterating over keys


stu = {"name":"Ritu","age":32,"grade":24}


for keys in stu.keys():
    print("Keys in stu dist: ",keys)
    
for values in stu.values():
    print("Values in stu dict: ", values)

for keys,values in stu.items():
    print("Elements in dict: ",keys,":",values)
    
for index ,keys in enumerate(stu):
    print(index,keys,stu[keys])
    
    
    
```

```python
## Nested Distionaries


students = {
    "student1": {"name": "Ritu", "age": 22, "course": "Python"},
    "student2": {"name": "Raj", "age": 21, "course": "Django"}
}

#1. Access data by keys
print(students["student1"]["name"])  
print(students["student2"]["name"])


#2. Loop through nested dictionary

for stu_id,info in students.items():
    print(f"{stu_id}:")
    
    for keys, values in info.items():
        print(f"{keys}:{values}")
```

```python
### Examples

emp ={
    "emp1":{"name":"Ritu","salary":400,"address":"Mohali"},
    "emp2":{"name":"Raj","salary":400,"address":"Chandigarh"}
}
print(emp.items())

print(emp["emp1"]["name"])
print(emp["emp2"]["name"])



for emp_id,emp_info in emp.items():
    print(f"{emp_id}")
    
    for keys,values in emp_info.items():
        print(f"{keys} :{values}")
```

```python
#### Dictionary Comhrehension

squares = {x:x**2 for x in range(5)}
print(squares)

#### CONDTIONAL DICT Comhrehension
sq_even = {x:x**2 for x in range(10) if x%2==0}
print(sq_even)
```
