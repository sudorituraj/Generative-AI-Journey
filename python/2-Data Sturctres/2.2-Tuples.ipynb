```python
#### ACCESSING TUPLE ELEMENTS

numbers = (1,2,3,4,5,6)

print("Index:",numbers[2])
print("Index:",numbers[0])
print("INdex:",numbers[-1])

## SLICING
print("Slicing")
print("Sliced:",numbers[2:])
print("Slced: ",numbers[::2])
print("sliecd: ",numbers[::-2])
```

```python

## TUPLES OPERATIONS

# * Concatnation 

Concationtion_tuple =numbers + mix_tpl
print(Concationtion_tuple)
```

```python
## TUPLES METHODS

numbers = (1,2,3,4,3,4)
print(numbers.count(3))
print(numbers.index(4))
```

```python
#### Packing and Unpacking tuples


### Packing tuples

packed_tuple = 1,"Hello",3.14
print("Packed Tuples: ", packed_tuple)


# UNpacked tuple

a,b,c = packed_tuple
print(a)
print(b)
print(c)
```

```python
# UNpacked with *

numbers = (1,2,3,4,5,6)

first,*middle,last = numbers
print(first)
print(middle)
print(last)




```

```python
text = ("Apple","bab","ritu","cat")

first,*middle,last = text 

print(middle)
print(first)
print(last)
```

```python


#### Nested Tuples
#### Nested List


lst = [[1,2,3,4],[6,7,8,9],[1,"Hello",3.14,"c"]]
print(lst[0][2])
print(lst[0][0:3])

```
