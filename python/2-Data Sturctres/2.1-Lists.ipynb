
#### Method      Description

* List Support concatentntaion 

1.append()	   - Adds an element at the end of the list

2.clear()	   -     Removes all the elements from the list

3copy()	     -   Returns a copy of the list

4.count()	  -      Returns the number of elements with the specified value

5.extend()	  -  Add the elements of a list (or any iterable), to the end of the current list

6.index()	 -       Returns the index of the first element with the specified value

7.insert()	   - Adds an element at the specified position

8.pop()	    -    Removes the element at the specified position

9.remove()	-    Removes the item with the specified value

10.reverse()	 -   Reverses the order of the list

```python
fruits = ["Apple","Mango","Palm","cherry","Gauva"]

fruits.append("orange")
print(fruits)
fruits.remove("Palm")
print(fruits)
fruits.pop() #last value romove
print(fruits)

index = fruits.index("Apple")
print("apple index:",index)

fruits.sort()
print("sorted fruits: ",fruits)

fruits.reverse()
print("reverse order:", fruits)

```

#### List COmprehension

1. Basic Syntax - [expression for items in iterable]

2. with conditional Logic- [repression for item in iterable if condtion]

example #[x**2 for x in range(15)]


3. NEster List comprehension -  [expression for item1 in iterable1 for item2 in iterable2]

```python
####  Basic List COmprehension


square = [num**2 for num in range(13)]

print("Squared between 1 to 13 is: ",square)

```

```python
### LIST COMpreshension with condition

lst = []
for i in range(10):
    if i%2 == 0:
        print(i)

```

```python
even_number = [num for num in range(10) if num%2 == 0]
print("Even number between 1-16 are: ",even_number)
```

```python
## NESTER LIST COMPREHENSION 

lst1 = [1,2,3,4]
lst2 = ["a","b","c","d"]
    
    
pair = [(i,j)for i in lst1 for j in lst2]

print(pair)
    
    
    
```
